## Label Encoding (Converting the labels into numeric form)

```python
# To import the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

## Label Encoding of Breast Cancer Dataset

```python
# Loading the data from csv file to pandas DataFrame
cancer_data = pd.read_csv("breast_cancer_data.csv")
```

```python
# To print the first 5 rows of the DataFrame
cancer_data.head()
```

```python
# To find the count of different labels 
cancer_data["diagnosis"].value_counts()
```

```python
# To Load the Label Encoder Function
label_encode = LabelEncoder()
```

```python
# To convert the value of B & M in diagnosis label
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# To Append the labels to the DataFrame
cancer_data["target"] = labels
```

```python
cancer_data.head()
```

# 0 -> Benign & 1 -> Malignant

```python
cancer_data["target"].value_counts()
```

## Label Encoding of the Irish Data

```python
# Loading the data from csv file to pandas DataFrame
iris_data = pd.read_csv("iris_data.csv")
```

```python
iris_data.head()
```

```python
iris_data["Species"].value_counts()
```

```python
# To Load the label Encoder 
iris_label_encoder = LabelEncoder()
```

```python
iris_labels = iris_label_encoder.fit_transform(iris_data.Species)
```

```python
iris_data["target"] = iris_labels
```

```python
iris_data.head()
```

```python
iris_data["target"].value_counts()
```

## Iris-setosa -> 0
## Iris-versicolor -> 1
## Iris-virginica -> 2
